```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
#color = sns.color_palette()
import datetime as dt
import scipy 


df = pd.read_csv("sample2012new.csv")
xg = pd.read_csv('xgb_starter2012.csv')
print(df.shape)
print(df.head(2))

print(xg.shape)
print(xg.head(2))
```

```python
xgw=xg.WHmodel.values
dfw=df.WHbuoy.values
ewb=dfw-xgw

print('mean absolute error xgb =',abs(ewb).mean())
print('xgb stats:',scipy.stats.describe(ewb))

date=df.date.values

date1 = pd.to_datetime(df.date)
```

```python
plt.figure(figsize=(18,6))
fontsize=20
labelsize=18
start=3000

plt.plot(date1[start:],xgw[start:],'r')
plt.plot(date1[start:],dfw[start:],'k.')

plt.tick_params(axis='both', which='minor', labelsize=labelsize)
plt.tick_params(axis='both', which='major', labelsize=labelsize)
```
